# **Filtrando o Dataframe por colunas, definindo X e Y e aplicando a escala**

```python
# COLUNAS A SEREM UTILIZADAS PARA PREVIS√ÉO
df_columns = ['LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']
```

```python
# SEPARANDO X E Y PARA PREVIS√ÉO
X = df[df_columns]
Y = df['default payment next month']
```

```python
# VISUALIZANDO O X
X
```

```python
# VISUALIZANDO O Y
Y
```

```python
# PRINTANDO O NOME DOS ATRIBUTOS E CALCULANDO OS TAMANHOS DOS DATAFRAMES DE X E Y
print('Nomes dos Atributos: ', df_columns)
print('Tamanho de X: ', X.shape)
print('Tamanho de Y: ', Y.shape)
```

```python
# APLICANDO A ESCALA A X
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# VISUALIZANDO O DATAFRAME DE X J√Å ATUALIZADO COM AS ESCALAS
X_scaled = pd.DataFrame(X_scaled, columns =X.columns)
X_scaled
```

# **Realizando as previs√µes e obtendo os primeiros resultados**

```python
# CALCULANDO A QUANTIDADE DE DADOS DE TREINO E TESTE
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=123)
```

```python
# PRINTANDO OS DADOS DE TREINO E TESTE
print('Tamanho de X_train: ', X_train.shape)
print('Tamanho de X_test: ', X_test.shape)
print('Tamanho de Y_train: ', Y_train.shape)
print('Tamanho de Y_test: ', Y_test.shape)
```

```python
# TREINANDO O MODELO DE REGRESS√ÉO LOG√çSTICA
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train, Y_train)
```

```python
# CALCULANDO A RELEV√ÇNCIA DE CADA COLUNA
for feature, coef in zip(df_columns, clf.coef_[0].tolist()):
    print(f"{feature}: {round(coef, 10)}")

print(f"Constante: {round(clf.intercept_[0], 10)}")
```

```python
# COMPARANDO OS DADOS DE TREINO AOS DADOS DE TESTE
from sklearn.metrics import accuracy_score

y_train_true = Y_train
y_train_pred = clf.predict(X_train)
y_test_true = Y_test
y_test_pred = clf.predict(X_test)

print(f"Acur√°cia de Treino: {round(accuracy_score(y_train_true, y_train_pred)*100, 2)}%")
print(f"Acur√°cia de Teste: {round(accuracy_score(y_test_true, y_test_pred)*100, 2)}%")
```

```python
# FAZENDO A MATRIZ DE CONFUS√ÉO DOS PRIMEIROS RESULTADOS

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_train_true, y_train_pred).plot()
ConfusionMatrixDisplay.from_predictions(y_test_true, y_test_pred).plot()
```

# **Realizando novas previs√µes com ajustes**

```python
# APLICANDO THRESOLD AO TREINO
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, Y_train)

y_pred_prob1 = logistic_model.predict_proba(X_train)[:, 1]

threshold = 0.33

y_pred_adjusted_train = (y_pred_prob1 >= threshold).astype(int)

print("Matriz de Confus√£o - Regress√£o Log√≠stica (Threshold Ajustado):")
print(confusion_matrix(y_train_true, y_pred_adjusted_train))
print("Relat√≥rio de Classifica√ß√£o - Regress√£o Log√≠stica (Threshold Ajustado):")
print(classification_report(y_train_true, y_pred_adjusted_train))
```

```python
# APLICANDO THRESOLD AO TESTE
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_test, Y_test)

y_pred_prob = logistic_model.predict_proba(X_test)[:, 1]

threshold = 0.33

y_pred_adjusted = (y_pred_prob >= threshold).astype(int)

print("Matriz de Confus√£o - Regress√£o Log√≠stica (Threshold Ajustado):")
print(confusion_matrix(y_test_true, y_pred_adjusted))
print("Relat√≥rio de Classifica√ß√£o - Regress√£o Log√≠stica (Threshold Ajustado):")
print(classification_report(y_test_true, y_pred_adjusted))
```

```python
# MATRIZ DE CONFUS√ÉO DO TREINO

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_train_true, y_pred_adjusted_train).plot()
```

```python
# MATRIZ DE CONFUS√ÉO DO TESTE

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test_true, y_pred_adjusted).plot()
```

```python
print(f"Acur√°cia de Treino: {round(accuracy_score(y_train_true, y_pred_adjusted_train)*100, 2)}%")
print(f"Acur√°cia de Teste: {round(accuracy_score(y_test_true, y_pred_adjusted)*100, 2)}%")
```

# **Realizando previs√µes usando Undersampling**

```python
df
```

```python
contagem = df['default payment next month'].value_counts().sort_index()
contagem
```

```python
contagem.argmin(), contagem.min()
```

```python
classes = [0, 1]
dicionario = {i: df[ df['default payment next month'] == i] for i in classes}
```

```python
for key in dicionario:
  dicionario[key] = dicionario[key].sample(contagem.min(), replace=False)
```

```python
df2 = pd.concat([dicionario[i] for i in classes], ignore_index=True, sort=False)
df2
```

```python
df2['default payment next month'].value_counts()
```

```python
X_u = df2[df_columns]
Y_u = df2['default payment next month']
```

```python
scaler = StandardScaler()
X_u_scaled = scaler.fit_transform(X_u)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_u_train, X_u_test, y_u_train, y_u_test = train_test_split(X_u_scaled, Y_u, test_size=0.2, random_state=42)
clf = LogisticRegression(max_iter=1000)
clf.fit(X_u_train, y_u_train)
y_pred_train_u = clf.predict(X_u_train)
print(classification_report(y_u_train, y_pred_train_u))
```

```python
ConfusionMatrixDisplay.from_predictions(y_u_train, y_pred_train_u).plot()
```
